```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/mypc/OneDrive/Desktop/Matplotlib/practice/Titanic.csv")
print(df.head())

df = df.dropna(subset=["Age",'Embarked',])

surivival_count = df['Survived'].value_counts()
plt.bar(['died','Survived'],surivival_count.values,color=['green','orange'],edgecolor='black',label='Died and Survived')
plt.title("Survival count of titanic ")
plt.ylabel("NUmer of passenger")
plt.savefig('Died_survived_in_titanic.png')
plt.legend()
plt.show()
# Group by Sex and sum Survived (number of survivors)
survival_gender = df.groupby("Sex")["Survived"].sum()

# Survival by Gender
plt.bar(survival_gender.index, survival_gender.values, color=['teal','red'], edgecolor='black',label='Survived by Gender')
plt.title("Number of Survivors by Gender")
plt.xlabel("Gender")
plt.ylabel("Survivors")
plt.savefig('survival_gender.png')
plt.legend()
plt.show()

# histogram with Survived and Died Ages
survival_gender = df[df['Survived']==1]['Age']
died_gender = df[df['Survived']==0]['Age']
plt.hist(survival_gender,bins=10,color='purple',alpha=0.6,label='Survived',edgecolor='black')
plt.hist(died_gender,bins=10,color='orange',alpha=0.6,label='Died',edgecolor='black')
plt.title("Survived passenger ages")
plt.xlabel("Age of passengers")
plt.ylabel("Count of passenger")
plt.legend()
plt.savefig('Survived and died.png')
plt.show()

plt.hist(df['Age'], bins=20, color='orange', edgecolor='black',label='Age Distribution')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.savefig('Age_distribution.png')
plt.legend()
plt.show()


pcclass = df['Pclass'].value_counts().sort_index()

plt.bar(pcclass.index,pcclass.values,color="red",edgecolor='black')
plt.title("passenger by pcclass")
plt.xlabel("Passenger class")
plt.savefig('pclass.png')
plt.show()
```
